Norint pasikartoti tam tikro dalyko sąvokas, pasileidžiame per Python tam skirtos funkcijos aprašymą

```python
import random
import sys

def kartojam_fiziką(words, reverse=False):
    if reverse:
        words = {y:x for x, y in words.items()}
    lt_meanings = list(words.keys())
    round_id = 1
    print('Paspausk ENTER, jei nusibodo')
    while len(lt_meanings)>0:
        random.shuffle(lt_meanings)
        your_score, total_score = 0, len(lt_meanings)
        print('-'*(9+len(str(round_id))))
        print(f'|ROUND: {round_id}|')
        print('-'*(9+len(str(round_id))))
        for w in lt_meanings.copy():
            r = input(f'/SĄVOKA: {w}|')
            if r == words[w].lower(): 
                print('\\Teisingai!')
                lt_meanings.remove(w)
                your_score += 1
            elif r == '': 
                print('*** Eik ilsėtis, tinginy')
                sys.exit()
            else: print('\\Neteisingai!', f'Turėjo būti: {words[w]}') 
        print(f'ATSAKEI: {your_score}/{total_score}, LIKO: {total_score - your_score}')
        round_id += 1       
    print('GERAI PAVAREI!')
```

Juos sieja formulė $n = \displaystyle \frac{\sin \alpha}{\sin \beta}$, čia $n$ yra lūžio rodiklis, lygus 1.33, kai spindulys sklinda vandeniu.

Remiantis Dekarto samprotavimu, neįmanoma nustatyti pozicijų, kuriose yra spalvos. Buvo aišku tik, kur jos yra, ir kokia kryptimi sklinda spindulys, tačiau to nepakako norint paaiškinti, iš kur atsiranda spalvos. Jis savo darbe pats tai pripažino. Jam pritrūko atrasti, kad rodiklis $n$ yra nevienodas skirtingų spalvų spinduliams. Rodiklio $n$ reikšmė lygi 1.344, kai spindulys violetinis ir 1.332, kai raudonas. Antroje nuorodoje yra pristatomi samprotavimo metodai, kuriuos Dekartas pristato kaip tinkamesnius naudoti darant bet kurį mokslinį tyrimą, nei lig tol nusistovėję ankstesnių mokslininkų darbuose.

```python
import matplotlib.pyplot as plt
import numpy as np
n = 1.333
degrees = np.arange(0, 100, 10)
radians = np.radians(degrees)
out = np.degrees(np.arcsin(np.sin(radians) / n))

fig, ax = plt.subplots(figsize=(15,10))
ax.plot(degrees, out, marker='o', color='k')
plt.title('Kokį lūžio kampą atitinka kritimo kampas?')
plt.xlabel('Kritimo kampas $\\alpha$, $^o$', fontsize=15)
plt.ylabel('Lūžio kampas $\\gamma$, $^o$', fontsize=15)

for x, y in zip(degrees[1:], out[1:]):
    plt.text(x,y, round(y, 2), dict(ha='right', va='bottom', fontsize=13, color='indigo'))
    
ax.set_xticks(np.arange(0,100,10))
ax.set_xticks(np.arange(0,91,1), minor=True)
ax.set_yticks(np.arange(0,60,10))
ax.set_yticks(np.arange(0,51,1), minor=True)
ax.grid(color = 'green', linestyle = '--', linewidth = 0.5, which='major', alpha=0.8)
ax.grid(color = 'green', linestyle = '--', linewidth = 0.5, which='minor', alpha=0.2)
ax.set_xlim(-0.5,90.5)
ax.set_ylim(-0.5,50.5)
ax.set_aspect('equal')
plt.show()
```
