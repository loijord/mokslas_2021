# The most frequent ways to create a numpy array

```python
import numpy as np
```

Create an array

```python
arr = np.array([[1, 3, 4, 7], [7, 8, 9, 23], [9, 7, 6, 5]])
arr
```

Create an array of the desired data-type

```python
arr = np.array([[1, 3, 4, 7], [7, 8, 9, 23], [9, 7, 6, 5]], dtype=np.uint8)
arr
```

Create an array of zeros with the same shape and type as a given array

```python
np.zeros_like(arr)
```

Create a new array of given shape and type, filled with zeros

```python
np.zeros((3, 4), dtype=int)
```

Create a new array of given shape and type, filled with ones

```python
np.ones((3, 4), dtype=bool)
```

Create a new array of given shape and type, without initializing entries

```python
with np.printoptions(linewidth=100, precision=5):
    print(np.empty((3, 4), dtype=float)) #rezervuojamos atsitiktinės atminties ląstelės float tipo duomenims
```

Create a new array of given shape and type, filled with `fill_value`

```python
np.full((3, 4), fill_value=2, dtype=np.uint8)
```

Create a 2-D array with ones on the diagonal and zeros elsewhere

```python
np.eye(5, dtype=int)
```

Given input array and the **number of repetitions for each element**, repeat elements of an array

```python
np.repeat([0, 1, 2, 3, 4], [9, 3, 0, 7, 1])
```

Given the input array and the **number of its repetitions** along each axis, repeat elements of an array

```python
np.tile(np.arange(3), 4)
```

```python
np.tile(np.arange(5), (3, 4))
```

### Independent random variables

Return array of random integers from low (inclusive) to high (exclusive)

```python
arr = np.random.randint(1, 7, size=(6, 7))
arr
```

To illustrate the concept of independent random variables (for example, rolling a dice), let's plot data and its distribution

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].plot(arr.ravel(), '--bo')
ax[0].set_xlabel('Trial', fontsize='x-large')
ax[0].set_ylabel('Value', fontsize='x-large')
ax[0].set_title('Data', fontsize='xx-large')

labels, counts = np.unique(arr, return_counts=True)
ax[1].bar(labels, counts, align='center', fc='g', ec='k')
ax[1].set_xlabel('Value', fontsize='x-large')
ax[1].set_ylabel('Count', fontsize='x-large')
ax[1].set_title('Distribution of data', fontsize='xx-large')
plt.show()
```

### Dependent random variables

```python
mu, sigma = 1230, 200
arr = np.random.normal(mu, sigma, size=(50, 4))
arr
```

```python
arr = np.around(arr, -1).astype(int)
arr
```

To illustrate the concept of dependent random variables (for example, measuring the daily intakes of adult), let's plot data and its distribution

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 2, figsize=(20, 5))
ax[0].plot(arr.ravel(), '--bo')
ax[0].set_xlabel('Trial', fontsize='x-large')
ax[0].set_ylabel('Value', fontsize='x-large')
ax[0].set_title('Data', fontsize='xx-large')

counts, labels = np.histogram(arr, bins=10, density=True)
ax[1].hist(arr.ravel(), bins=labels, fc='g', ec='k')
ax[1].set_xticks(labels)
ax[1].set_xlabel('Value', fontsize='x-large')
ax[1].set_ylabel('Count', fontsize='x-large')
ax[1].set_title('Distribution of data', fontsize='xx-large')
plt.show()
```
