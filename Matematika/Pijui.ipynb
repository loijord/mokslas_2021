## Eksperimentas 1

Skritulio ploto įvertinimas - puikus pavyzdys, kokias priemones galima panaudoti norint rasti plotą figūros, turinčios kreivą paviršių. Pamėginkime virtualiai atlikti tokį eksperimentą. Kiekvienam skaičiui $n \ge 4$ padalinkime kvadratą į $n \times n$ smulkesnių kvadratėlių ir užtušuokime tuos, kurių centrai yra įbrėžto skritulio viduje.

* Žvilgterėkite į paveikslėlius, kur $n$ lygus 4, 7, 12, 22. Kiek langelių užima kvadratas? Kiek langelių užima užspalvintas regionas?
* Kaip galima apytiksliai apskaičiuoti skritulio plotą langeliais?
* Ar skaičiavimo tikslumas gerėja, kai $n$ didėja?
* Kurią dalį visų langelių sudaro užspalvinti langeliai, kai $n$ lygus 4, 7, 12, 22? 
* Ar šita dalis daug keičiasi, kai $n$ yra didelis?

Žemiau yra pateikiamas Python kalbos kodas, kurį paleidus galima virtualiai atlikti kvadrato pjaustymus su nedideliais $n$

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import numpy as np

def vaizduok_skritulius(a, b, start=4, print=False):
    fig, ax = plt.subplots(a, b, figsize = (b*10, a*10))
    stats = []
    for idx in range(start, start+a*b):
        i, j = divmod(idx - start, b)
        r = idx/2
        center = (r, r)
        x, y = np.indices((idx, idx))
        s = np.hypot(x - center[0] + 0.5, y - center[1] + 0.5)<r

        suma, viso = np.sum(s), s.size
        rezultatas = np.round(suma/viso, 3)
        stats.append(rezultatas)
        
        if print: print(f'Kai kvadratas yra {len(x)}*{len(y)}, tai užtušuota {suma}/{viso} = {rezultatas}')
        
        ax[i, j].xaxis.set_tick_params(labeltop=True)
        ax[i, j].xaxis.set_tick_params(labelbottom=False)

        ax[i, j].pcolormesh(s, edgecolors='k', linewidth=2, cmap='Greens')
        ax[i, j].set_xticks(np.arange(idx+1))
        ax[i, j].yaxis.set_major_locator(mticker.FixedLocator(np.arange(idx+1)))
        ax[i, j].set_yticklabels(np.arange(idx, -1, -1))
        
        ax[i, j].add_patch(plt.Circle(center, radius=r, linewidth=1, ec='k', fc=(1, 1, 1, 0.8)))
    plt.show()
    fig.savefig('skrit', bbox_inches='tight', facecolor='lavenderblush', edgecolor='k')
    return stats

out = vaizduok_skritulius(5, 5)
```

Kompiuteris skaičiavimus gali atlikti greit, todėl galima užduoti komandą pasižiūrėti, kokias užtušuotų kvadratėlių dalis duos pirmieji 300 pjaustymų.

```python
def skritulių_statistika(N, start=4):
    stats = []
    for idx in range(start, start+N):
        r = idx/2
        center = (r, r)
        x, y = np.indices((idx, idx))
        s = np.hypot(x - center[0] + 0.5, y - center[1] + 0.5)<r
        suma, viso = np.sum(s), s.size
        rezultatas = np.round(suma/viso, 5)
        stats.append(rezultatas)
    return stats
    
N, start = 300, 4
out = skritulių_statistika(N, start)

fig, ax = plt.subplots(figsize=(10, 10))
ax.set_xlabel('Kvadrato kraštinės ilgis')
ax.set_ylabel('Kurią dalį kvadrato sudaro užspalvinta dalis')
plt.plot((start, start+N), (np.pi/4, np.pi/4), c='r')
plt.plot(range(start, start+N), out)

plt.show()
```

Pasižiūrėję į šį grafiką galime atsakytį į paskutinį klausimą:
    
* *Ar šita dalis daug keičiasi, kai $n$ yra didelis?*

Atsakymas: kuo $n$ didesnis, tuo užtušuota dalis mažiau keičiasi. Panašu, kad ji artėja į skaičių 0.785. 

Tačiau tai tik prielaida. Niekas kol kas negali garantuoti, kad su $n$, didesniais už 300, neįvyks taip:

```python
fig, ax = plt.subplots(figsize=(20, 10))
ax.set_xlabel('Kvadrato kraštinės ilgis')
ax.set_ylabel('Kurią dalį kvadrato sudaro užspalvinta dalis')

fake_addend = (np.sin(2 *np.pi * np.arange(N)/100) + np.random.random(N)/2)* (1 - np.cos(2 *np.pi * np.arange(N)/N))/100
plt.plot((start, start+3*N), (np.pi/4, np.pi/4), c='r')
plt.plot(range(start, start+3*N), out + (np.pi/4 + fake_addend*0.5).tolist() + (np.pi/4 + fake_addend).tolist())

plt.show()
```

## Kitokia geometrija

***Debesys - ne rutuliai, kalnai - ne kūgiai, krantai - ne apskritimai, ir nei medžio žievė glodi, nei žaibas tiesėm sklinda*** - taip rašė IBM bendradarbis, JAV profesorius B.B. Mandelbrotas savo 1977 metais išleistoje knygoje *Gamtos fraktalų geometrija*. Tais laikais tai buvo pirmoji visiems prieinama knyga, paskatinusi atkreipti dėmesį į egzotinę geometrijos rūšį - fraktalų geometriją. Fraktalas, supaprastintai tariant, yra kreivė, panaši į save. Nors tokios kreivės koncepcija buvo pradėta vystyti dar prieš šimtmetį nuo šios knygos atsiradimo, tačiau anksčiau idėjos apie tokias kreives nuguldavo į plačiajai visuomenei sunkiai prieinamus mokslininkų straipsnius.

Žvilgterėkime į vienos tokios kreivės pavyzdį:

![](koch.png)

Matome, kad tokios figūros plotą suskaičiuoti remiantis ankstesniais būdais būtų išties nelengva. Dar viena įdomybė: šios figūros perimetras sulig kiekvienu žingsniu tampa vis didesnis ir jo didėjimui nėra galo (**begalinio perimetro savybė**)

Kai kurių figūrų netaisyklingas kreivumas buvo žinomas ir anksčiau, prieš ne vieną tūkstantį metų. Pavyzdžiui pjaustant tam tikrą erdvinį geometrinį kūną - kūgį [galima atrasti](https://cseligman.com/text/history/ellipses.htm) net keturių skirtingų tipų kreives: skritulį, elipsę, parabolę ir hiperbolę

![](conic.png)

Mandelbroto knygoje buvo kritikuojamas mokyklinės (elementariosios) geometrijos ribotumas ir sausumas. Mokyklinė geometrija nėra pajėgi atvaizduoti debesų, kalnų, krantų ar medžio formų. Daugybė gamtos formų yra tokios netaisyklingos ir padrikos, kad palyginus su elementariąja geometrija, gamta demonstruoja ne tik aukštesnį, bet ir visiškai kitokį sudėtingumo laipsnį. Tai sukėlė revoliuciją daugybėje sričių, iš esmės susijusių su gamtos modeliavimu. 1977m. pakankamai galingų skaičiavimo mašinų, galinčių atvaizduoti knygoje *Gamtos fraktalų geometrija* aprašomus fraktalus, buvo nedaug. Tačiau skaičiavimo pajėgumams greitai augant, teprireikė vos kelių metų, kad modernios kompiuterinės grafikos pagalba taptų įmanoma vizualizuoti mokslininkų atrastas kreives ir atskleisti jų grafikos nepaprastą grožį:

![](fracta.png)

Atrodžiusi vien loginiu ir estetiniu užsiėmimu fraktalų teorija apie 2000 m. susilaukė ir konkretaus pritaikymo - mažinant mobiliųjų telefonų dydį buvo pasiūlyta telefono viduje montuoti fraktalo pavidalo antenas, kurios pasirodė ne mažiau efektyvios už išorines (pagal **begalinio perimetro savybę**).
